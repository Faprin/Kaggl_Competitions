### Creación de nuevas variables a partir de los datos

# 1. Primera aproximación: Regresión Logística

```python
df_limpio.head()
```

Como tenemos aun algunas variables que son de tipo string, vamos a pasarlas por un onehot encoder para que el modelo lo entienda bien

```python
# esta es la forma de hacer one_hot de manera rápida
df_limpio = pd.get_dummies(df_limpio, ['Sex','Embarked'], drop_first=True)
df_limpio.head()
```

```python
X = df_limpio.drop('Survived', axis=1)
y = df_limpio['Survived']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = pd.DataFrame(
    scaler.fit_transform(X_train),
    columns=X_train.columns,
    index=X_train.index
)

X_test_scaled = pd.DataFrame(
    scaler.transform(X_test),
    columns=X_test.columns,
    index=X_test.index
)
```

🚨 La Regla de Oro del Escalado: > El modelo es como un alumno haciendo un examen. Nunca puede ver información de los datos de test (X_test) antes de tiempo. Por eso, el escalador "estudia" (calcula la media y desviación estándar) solo con el X_train. Luego, aplica esa misma regla a ambos conjuntos.

```python
from sklearn.linear_model import LogisticRegression

LR = LogisticRegression().fit(X_train_scaled, y_train)
```

```python
y_hat_lr = LR.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

acc_score = accuracy_score(y_hat_lr, y_test)
print(f"Acc score: {acc_score}")

print("\n---- Confussion Matrix ----")
print(confusion_matrix(y_hat_lr, y_test))

print("\n---- Classification Report ----")
print(classification_report(y_hat_lr, y_test))
```

### Interpretación de la matriz de confusión
| | Dijo que MUERE (0) | Dijo que VIVE (1) |
| :--- | :--- | :--- |
| **Realmente MURIÓ (0)** | Verdaderos Negativos (TN) | Falsos Positivos (FP) |
| **Realmente VIVIÓ (1)** | Falsos Negativos (FN) | Verdaderos Positivos (TP) |

### Pruebas de diagnóstico
#### Prueba de overfitting o subfitting

```python
train_score_lr = LR.score(X_train_scaled, y_train)
print(f"Train score: {train_score_lr}")
```

#### Prueba de Equidad
Vemos que el recall no varía mucho, por lo que el modelo no se ha vuelto pesimista y no es necesario hacer un class_weight balanceado

#### Prueba de las características que más afectan

```python
coeficientes_lr = LR.coef_[0]

df_importancia = pd.DataFrame({
    'Variables': X_train.columns,
    'Importancia': coeficientes_lr
})
# Calculamos el valor absoluto para ordenar por "Impacto Total" (sea positivo o negativo)
df_importancia['Impacto_Absoluto'] = abs(df_importancia['Importancia'])
df_importancia = df_importancia.sort_values(by='Impacto_Absoluto', ascending=False)

print("--- LO QUE MÁS PESO TIENE PARA EL MODELO ---")
print(df_importancia[['Variables', 'Importancia']])
```

```python
df_resultados_lr = X_test.copy()
df_resultados_lr['Realidad'] = y_test
df_resultados_lr['Predicciones'] = y_hat_lr
df_errores = df_resultados_lr[df_resultados_lr['Realidad'] != df_resultados_lr['Predicciones']]
# df_errores
```

# 2. Segunda aproximación: RandomForest

No voy a escalar los datos ya que el algoritmo de los árboles dará el mismo resultado. Sin embargo, necesito que las variables sigan codificadas con one_hot. Aunque ahora podemos dejar que el algoritmo de one_hot cree todas las categorías necesarias

```python
X.head()
```

```python
X_train.head()
```

Como veo, ya esta en one_hot y no hace falta esclarlo, tengo los dataset listos

```python
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(n_estimators=100, random_state=54, class_weight='balanced',
                           max_depth=6,
                           min_samples_split=10,
                            min_samples_leaf=5).fit(X_train, y_train)
```

```python
y_hat_rf = RF.predict(X_test)
```

```python
acc_score = accuracy_score(y_hat_rf, y_test)
print(f"Acc score: {acc_score}")
print("\n---- Matriz de Confusion ----")
print(confusion_matrix(y_hat_rf, y_test))
print("\n---- Classification report ----")
print(classification_report(y_hat_rf, y_test))
```

### Pruebas de Diagnóstico
#### Prueba de fit

```python
train_score_rf = RF.score(X_train, y_train)
print(f"Train score: {train_score_rf}")
print("Test score:", RF.score(X_test, y_test))
```

Tengo overfitting. Si nos fijamos en la diferencia abismal entre el acc score y el train score, vemos que el train score es mucho mayor al acc. Por lo tanto, nos encontramos ante una situación de overfitting

Por lo tanto, vamos a aplicar las siguientes soluciones:
- max_depth: número de preguntas que hará un arbol (profundiadd)
- min_samples_split: número de ejemplos que necesitará como  mínimo para poder abrir una nueva rama
- min_samples_leaf: número de ejemplos que necesitará para  poder tomar una decisión final

#### Ahora nos fijamos en las variables que más ha usado este algoritmo

```python
importancias_rf = RF.feature_importances_
df_importancia_rf = pd.DataFrame({
    'Variables': X_train.columns,
    'Importancias': importancias_rf
})

df_importancia_rf = df_importancia_rf.sort_values(by='Importancias', ascending=True)
plt.figure(figsize=(10,8))
plt.barh(df_importancia_rf['Variables'], df_importancia_rf['Importancias'], color='teal')
plt.show()
```

Embarked es puro ruido, sin embargo familia, quizás el número confunda al modelo, por lo tanto lo voy a agrupar en familias grandes, medianas o pequeñas. 

```python
df_limpio.head()
```

```python
def fam_transform(n):
    if n == 1:
        return 'Solo'
    if 2 <= n <= 4:
        return 'Mediana'
    if n >4:
        return 'Grande'

df_limpio['Family_tam'] = df_limpio['Family_size'].apply(fam_transform)
```

```python
# 2. Agrupar las edades (Binning con Pandas)
bins = [0, 12, 25, 60, 100]
etiquetas = ['Nino', 'Joven', 'Adulto', 'Anciano']
df_limpio['Rango_Edad'] = pd.cut(df_limpio['Age'], bins=bins, labels=etiquetas)
```

```python
df_limpio = df_limpio.drop(['Embarked_Q', 'Embarked_S', 'Family_size', 'Age'], axis=1)
df_limpio = pd.get_dummies(df_limpio, columns=['Rango_Edad', 'Family_tam'], drop_first=True)
df_limpio.head()
```

# Tercera aproximación: XGBoost

Las reglas de oro para alimentar a XGBoost

Para sacar la máxima nota posible en el Titanic, prepara tus variables así:

    Números crudos (Sin agrupar): Mantén Age, Fare y Family_size como números exactos. No hagas binning.

    One-Hot Encoding (Obligatorio): Aunque XGBoost es muy listo, la envoltura de Python sigue necesitando que las letras sean números. Tienes que usar pd.get_dummies(drop_first=True) para Sex y Embarked.

    Sin Escalar (StandardScaler): Igual que el Random Forest, a XGBoost le da igual si la tarifa es 70 o 0.85. Ahórrate el escalador.

    Valores Nulos (NaNs): A diferencia de todos los modelos anteriores, XGBoost sabe qué hacer con los huecos vacíos por defecto. Sin embargo, como ya hiciste el trabajo duro de rellenar la edad con la media, puedes dejarlo así.

```python
df_xgboost = df.copy()
df_xgboost.head()
```

```python
df_xgboost = df_xgboost.drop(['Name', 'Ticket', 'PassengerId', 'Cabin'], axis=1)
df_xgboost['Age'] = df_xgboost['Age'].fillna(df_xgboost['Age'].mean())
```

```python
df_xgboost.isna().sum()
```

```python
# elimino los nulos de embarkedabs
df_xgboost = df_xgboost.dropna()
```

```python
# agrupamos las variables subsp y parch para dejar los datos de familia en tipo numéricoabs
df_xgboost['Family_size'] = df_xgboost['SibSp'] + df_xgboost['Parch'] + 1
df_xgboost = df_xgboost.drop(['SibSp', 'Parch'], axis=1)
```

```python
df_xgboost.head()
```

```python
# hacemos el encoding de las vareiables que son de tipo string
df_xgboost = pd.get_dummies(df_xgboost, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
df_xgboost.head()
```

```python
X_xgb = df_xgboost.drop('Survived', axis=1)
y_xgb = df_xgboost['Survived']
```

```python
X_xgb_train, X_xgb_test, y_xgb_train, y_xgb_test = train_test_split(X_xgb, y_xgb, test_size=0.3, random_state=42, )
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    eval_metric='logloss'  # Métrica interna para evaluar sus propios errores

).fit(X_xgb_train, y_xgb_train)
```

```python
xgb_pred = xgb.predict(X_xgb_test)

acc_xgb_score = accuracy_score(xgb_pred, y_xgb_test)

print(f"Acc score: {acc_xgb_score}")
print("Train score:", xgb.score(X_xgb_test, y_xgb_test))
```

```python
print("--- Confusion Matrix ----")
print(confusion_matrix(xgb_pred, y_xgb_test))
print("\n---- Classification report ----")
print(classification_report(xgb_pred, y_xgb_test))
```

## Como última aproximación tenemos que rescatar variables del dataset original
- Name contiene los pronombres de personas importantes:
- Cabin, tiene casi 700 filas en NA, pero eso significa que aquellos que no tenían sitio en cabin, no eran de primera clase
- Tiket

```python
df_xgb = df.copy()
```

```python
df_xgb.head()
```

```python
df_xgb['Title'] = df_xgb['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
titulos_raros = ['Lady', 'Countess','Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona']
df_xgb['Title'] = df_xgb['Title'].replace(titulos_raros, 'Rare')
df_xgb['Title'] = df_xgb['Title'].replace('Mlle', 'Miss') # Señorita en francés
df_xgb['Title'] = df_xgb['Title'].replace('Ms', 'Miss')
df_xgb['Title'] = df_xgb['Title'].replace('Mme', 'Mrs')   # Señora en francés

# 4. IMPUTACIÓN INTELIGENTE DE LA EDAD
# Rellenamos los nulos de 'Age' calculando la MEDIANA de edad para cada grupo de 'Title'
df_xgb['Age'] = df_xgb['Age'].fillna(df_xgb.groupby('Title')['Age'].transform('median'))
```

```python
df_xgb.isna().sum()
```

```python
df_xgb['Cabin'] = df_xgb['Cabin'].fillna('Sin_cabina')
df_xgb.head()
```

```python
df_xgb = df_xgb.drop(['SibSp', 'Parch', 'Ticket', 'Embarked', 'Name'], axis=1)
df_xgb_limpio = pd.get_dummies(df_xgb, columns=['Sex', 'Cabin', 'Title'], drop_first=True)
```

```python
X_xgb = df_xgb_limpio.drop('Survived', axis=1)
y_xgb = df_xgb_limpio['Survived']
```

```python
X_xgb.head()
```

```python
X_xgb_train, X_xgb_test, y_xgb_train, y_xgb_test = train_test_split(X_xgb, y_xgb, test_size=0.3, random_state=42, )

from xgboost import XGBClassifier
xgb = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    eval_metric='logloss'  # Métrica interna para evaluar sus propios errores

).fit(X_xgb_train, y_xgb_train)

xgb_pred = xgb.predict(X_xgb_test)

acc_xgb_score = accuracy_score(xgb_pred, y_xgb_test)

print(f"Acc score: {acc_xgb_score}")
print("Train score:", xgb.score(X_xgb_test, y_xgb_test))

print("--- Confusion Matrix ----")
print(confusion_matrix(xgb_pred, y_xgb_test))
print("\n---- Classification report ----")
print(classification_report(xgb_pred, y_xgb_test))
```
